```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv(r'C:\Users\Nita\Desktop\ProjectsPersonal\Ranking/movies_data.csv')
df.head()
```

### Popularity-Based Ranking:

```python
top_20_popularity = df.sort_values(by='popularity', ascending=False).head(20)
top_20_popularity
```

### Vote Count-Based Ranking:

```python
top_20_vote_count = df.sort_values(by='vote_count', ascending=False).head(20)
top_20_vote_count
```

### Vote Average-Based Ranking:

```python
top_20_vote_average = df.sort_values(by='vote_average', ascending=False).head(20)
top_20_vote_average
```

### Hybrid (Popularity and Vote Average) Ranking:

```python
df['popularity_vote_avg'] = df['popularity'] * df['vote_average']
top_20_hybrid = df.sort_values(by='popularity_vote_avg', ascending=False).head(20)
top_20_hybrid
```

### Weighted Rating:

```python
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.90)
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
df['weighted_rating'] = df.apply(weighted_rating, axis=1)
top_20_weighted = df.sort_values(by='weighted_rating', ascending=False).head(20)
top_20_weighted
```

### Recent Releases (Latest Movies) Ranking:

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
top_20_recent = df.sort_values(by='release_date', ascending=False).head(20)
top_20_recent
```

### Genre-Specific Ranking (e.g., Comedy):

```python
comedy_movies = df[df['genres'].apply(lambda x: 'Comedy' in str(x))]
top_20_comedy = comedy_movies.sort_values(by='popularity', ascending=False).head(20)
top_20_comedy
```

### Vote Count and Genre Combined Ranking:

```python
comedy_movies = df[df['genres'].apply(lambda x: 'Comedy' in str(x))]
top_20_comedy_highest_voted = comedy_movies.sort_values(by='vote_count', ascending=False).head(20)
top_20_comedy_highest_voted
```
